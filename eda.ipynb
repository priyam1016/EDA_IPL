NOW LET'S ANALYSE HOW TOSS IS IMPACTING GAME RESULTS

```python
#How toss affects a team chances of winning
len(matches[matches['toss_winner']==matches['winner']])/len(matches) * 100
#It seems there is fair 50-50 chance of both teams winning despite toss results
```

```python
# Let us see how a particular team gets effected with a toss win
temp1=matches[matches['toss_winner']==matches['winner']]
temp2=matches.groupby('toss_winner')['winner'].count()
temp1=temp1.groupby('toss_winner')['winner'].count()
temp=temp1/temp2*100
plt.xticks(rotation=90)
plt.plot(temp)
temp
"""Seems like Chennai Super Kings and Gujarat Lions have  higher probability of winning the match if they win the toss while Pune Warriors is most likely to loss a match if they win the toss"""
```

```python
# Do Toss is a deciding Factor in Finals?
len(Finals[Finals['toss_winner']==Finals['winner']])/len(Finals) * 100
```

We have seen 12 Finals so far and 9 out 12 have won by the team winning the toss which gives a percentage of 75..Well by seeing this it seems toss winning becomes crucial in Finals.
